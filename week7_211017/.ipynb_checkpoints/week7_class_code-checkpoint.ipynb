```python
import numpy as np
# 移动：
M = np.float32([[1,0,-300],[0,1,200]])
plt.imshow(img)
print(img.shape)
#print(img[0+50,0])
img_1=cv2.warpAffine(img,M,(1300,2200))

plt.imshow(img_1)
#print(img_1[200+50,300])
```

```python
# 图像的旋转变换与拉伸变换
```

```python
#pts1 = np.float32([[50,50],[200,50],[50,200]])
#pts2 = np.float32([[10,100],[200,50],[100,250]])
#M = cv2.getAffineTransform(pts1,pts2)
#print(M)
#180/3.14
theta=0.5
M = np.float32([[np.cos(theta),-np.sin(theta),100],[np.sin(theta),np.cos(theta),200]])
#M = np.float32([[0.1,0,100],[0,2,100]])
# 变换矩阵，平移，斜切，旋转
# affine
cols=800
rows=800
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

```python
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[100,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
print(M)
# 拉伸变换后者透视变换
dst = cv2.warpPerspective(img,M,(300,300))
plt.imshow(dst)
```

# 图像滤波/卷积

```python
kernel = np.ones((3,3),np.float32)/8
kernel=-kernel
kernel[0,:]=[-1,-1,-1]
kernel[1,:]=[0,0,0]
kernel[2,:]=[1,1,1]


print(kernel)
plt.imshow(img)
```

```python
#dst=cv.filter2D(src, ddepth, kernel[, dst[, anchor[, delta[, borderType]]]])；当ddepth=-1时，表示输出图像与原图像有相同的深度。
print(img.shape)
result = cv2.filter2D(img,-1,kernel)
result.shape
print(result[0,0])
plt.imshow(result*255)
```

```python
result = cv2.filter2D(result,-1,kernel)

plt.imshow(result)
result.shape
```
